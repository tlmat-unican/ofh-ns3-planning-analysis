```python
import numpy as np
import matplotlib.pyplot as plt  


folder = "./sim_results/"


plt.rcParams["figure.autolayout"] = True

Flow_min_value = 1
Flow_max_value = 6
Step = 1
configs_txt = ["RU.1", "RU.2", "RU.3", "RU.4", "RT11R3", "RT11R0"]
Configs = np.arange(Flow_min_value,Flow_max_value+Step,Step)
Nsim = 1
delay = []

result = "endToEndBackground"
for conf in Configs:
    foldername = "{}TxFile{}.log".format(folder,conf)
    with open (foldername) as datatoread:
        Tx = np.loadtxt(datatoread)
    foldername = "{}RxFile{}.log".format(folder,conf)
    with open (foldername) as datatoread:
        Rx = np.loadtxt(datatoread)

    # Ensure Tx has the same length as Rx
    min_length = min(len(Tx), len(Rx))
    delay.append(Rx[:min_length,1] - Tx[:min_length,1])

    # ThputAuxstream.append((np.sum(Rx[:,1])*8)/(Rx[len(Rx)-1,0]- Rx[0,0]))       
delay = np.array(delay)



# plt.ylabel('Delay (ms)')
# plt.xlabel('FLOW')



# Set a custom color palette
colors = plt.cm.Set3.colors

# Boxplot with enhanced style

bp = plt.boxplot(
    delay.transpose() * 1e3,
    labels=configs_txt,
    showmeans=True,
    showfliers=False,
    patch_artist=True,  # fill with color
    boxprops=dict(facecolor=colors[0], color=colors[9]),
    capprops=dict(color=colors[9]),
    whiskerprops=dict(color=colors[9]),
    flierprops=dict(markeredgecolor=colors[9], markerfacecolor=colors[9]),
    medianprops=dict(color=colors[9]),
    meanprops=dict(marker="o", markerfacecolor=colors[3], markeredgecolor=colors[3]),
)

# Customize labels and title
plt.ylabel("Delay (ms)")
plt.xlabel("FLOW")
plt.title("Delay Boxplots for Different Configurations")

# Add gridlines
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.grid()
plt.show()
```
